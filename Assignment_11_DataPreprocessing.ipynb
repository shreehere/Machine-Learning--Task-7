This Data set has 10 rows and 4 columns.
Column 1: Country -  Consists of 3 countries; France, Spain and Germany
Column 2: Age - The age is given for all rows except one row it is mentioned as NaN 
Column 3: Salary - Except 1 row all the other values are given in the dataset
Column 4: Purchased - This is categorical column with two values, 'Yes' and 'No'

**Step 1: Importing the libraries**

```python
import numpy as np
import pandas as pd
```

```python
#To handle the missing data
from sklearn.impute import SimpleImputer

#To encode categorical data
from sklearn.preprocessing import OrdinalEncoder

#To split the train and the test data
from sklearn.model_selection import train_test_split

#To standardize the data
from sklearn.preprocessing import StandardScaler
```

**Step 2: Importing dataset**

```python
df=pd.read_csv("/content/Data.csv")
df
```

**Step 3: Handling the missing data**

```python
#Using Simple Imputer to replace the Nan values in the data

imputer = SimpleImputer(strategy='mean', missing_values=np.nan)
imputer = imputer.fit(df[['Age','Salary']])
df[['Age','Salary']] = imputer.transform(df[['Age','Salary']])
df
```

```python
df.info()  #Now the data has all non-null values
```

**Step 4: Encoding categorical data**

```python
#Encoding the categorical data using OrdinalEncoder

encoding=OrdinalEncoder()
df['Purchased']=encoding.fit_transform(df[["Purchased"]])
```

```python
df  #Encoding only the column 'Purchased'
```

**Step 5: Creating a dummy variable**

```python
#creating dummy variable for the different countries in the 'Country' column using get_dummies method

df=pd.get_dummies(df)
df
```

**Step 6: Splitting the datasets into training sets and Test sets**

```python
#splitting the x set without the column 'Purchased' and keeping it for y

x=df.iloc[:,:-1]
y=df['Purchased']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30)
```

```python
x_test  #just to see the x_test values
```

**Step 7: Feature Scaling**

```python
scaling=StandardScaler()
x_train=scaling.fit_transform(x_train)
x_test=scaling.fit_transform(x_test)
```
