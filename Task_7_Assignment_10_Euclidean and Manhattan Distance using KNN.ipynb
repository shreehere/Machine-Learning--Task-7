##K- Nearest Neighbour (KNN) Classifier

```python
# importing all the necessary Libraries

import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report
from scipy.spatial import distance
```

Reading the train dataset in Pandas DataFrame using read_csv

```python
train_df=pd.read_csv("/content/trainKNN.txt",header=None)
train_df
```

Changing the column names to the original data column names

```python
train_df.columns=['Id','RI','Na','Mg','Al','Si','K','Ca','Ba','Fe','Type of glass']
train_df
```

Droping the 'Id' column since it is not required

```python
train_df=train_df.drop('Id',axis=1)
train_df
```

Doing the same for the test data

```python
test_df=pd.read_csv("/content/testKNN.txt",header=None)
test_df.columns=['Id','RI','Na','Mg','Al','Si','K','Ca','Ba','Fe','Type of glass']
test_df=test_df.drop('Id',axis=1)
test_df
```

```python
train_df.describe()
```

```python
test_df.describe()
```

Finding unique values for train and test data

```python
train_df['Type of glass'].unique()
```

```python
test_df['Type of glass'].unique()
```

```python
from sklearn.preprocessing import StandardScaler
```

**Splitting the test and the train Data**

```python
x_train=train_df.iloc[:,1:-1]
y_train=train_df['Type of glass']
x_test=test_df.iloc[:,1:-1]
y_test=test_df['Type of glass']
```

**Feature Scaling**

```python
st=StandardScaler()
x_train=st.fit_transform(x_train)
```

```python
x_test=st.fit_transform(x_test)
```

```python
x_train
```

```python
x_test
```

#Euclidean Distance

```python
euc_knn=KNeighborsClassifier(metric=distance.sqeuclidean)
```

```python
#model created for Euclidean distance

euc_knn.fit(x_train,y_train)
```

```python
y_prediction_euc=euc_knn.predict(x_test)
```

```python
y_prediction_euc
```

```python
from sklearn.metrics import accuracy_score
```

```python
#Finding accuracy level of the Euclidean prediction

accuracy_euc=accuracy_score(y_test,y_prediction_euc)*100
accuracy_euc
```

#Manhattan distance

```python
#model created for Manhattan distance

manhattan_knn=KNeighborsClassifier(metric='manhattan')
```

```python
manhattan_knn.fit(x_train,y_train)
```

```python
y_prediction_manhattan=manhattan_knn.predict(x_test)
```

```python
#Finding accuracy level of the Manhattan prediction

accuracy_manhattan=accuracy_score(y_test,y_prediction_manhattan)*100
accuracy_manhattan
```

```python
#Ignoring warnings 

import warnings
warnings.filterwarnings('ignore')
```

**Classification Report for Euclidean and Manhattan**

```python
print(classification_report(y_test,y_prediction_euc))
print("-----------------------------------------------------------------------")
print("The accuracy score of Euclidean Distance for the test Data :", accuracy_euc  )
```

```python
print(classification_report(y_test,y_prediction_manhattan))
print("-----------------------------------------------------------------------")
print("The accuracy score of Manhattan Distance for the test Data :", accuracy_manhattan)
```

#CONCLUSION

The accuracy score of Manhattan Distance is 66.66 and the accuracy for Euclidean score is 61.11. This shows the model created for Manhattan Distance fits more in the given data
